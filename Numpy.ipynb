# Advanced Numpy Library Function

```python
#Numpy array vs Python list
a = [i for i in range (10000000)]
b = [i for i in range (10000000,20000000)]
import time
start = time.time()
c = [a[i] + b[i] for i in range(len(a))]
print(time.time()-start)
```

```python
import numpy as np
a = np.arange(10000000)
b = np.arange(10000000,20000000)
start = time.time()
c=a+b
print(time.time()-start)
```

**Numpy operations are faster than Python list**

```python
import sys
a = [i for i in range (10000000)]
sys.getsizeof(a)
```

```python
b = np.arange(10000000,dtype=np.int16)
sys.getsizeof(b)
```

```python
#sigmoid
def sigmoid(array):
  return 1/(1+np.exp(array))
```

```python
a = np.arange(10)
sigmoid(a)
```

```python
#mean sqare error
actual = np.random.randint(1,100,50)
predicted = np.random.randint(1,100,50)
def mse(actual,predicted):
  return np.mean((actual-predicted)**2)
mse(actual,predicted)
```
